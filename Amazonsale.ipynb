```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html

# Load dataset
df = pd.read_csv(r'C:\Users\sayali\Documents\Amazon_sales_report2.csv')
df.dropna(inplace=True)
df.rename(columns=lambda x: x.strip(), inplace=True)
df['Date'] = pd.to_datetime(df['Date'])

# Ensure 'Courier Status' and 'ship-state' are cleaned
df['Courier Status'] = df['Courier Status'].str.strip()
df['ship-state'] = df['ship-state'].str.strip()

# Define color map for Courier Status
color_map = {
    'On the Way': '#FF5733',
    'Shipped': '#335BFF',
    'Unshipped': '#33FF57'
}

# Aggregate data for Shipping Status Chart
df_shipping_status = df.groupby("Courier Status", as_index=False)["Amount"].sum()

# Filter for Fulfilment type
df_filtered = df[df['Fulfilment'].isin(['Amazon', 'Merchant'])]

# Get Top 10 States for distribution (Sorted in descending order)
top_states = df['ship-state'].value_counts().nlargest(10).index
df_top_states = df[df['ship-state'].isin(top_states)].groupby('ship-state', as_index=False)['Amount'].sum()
df_top_states = df_top_states.sort_values(by='Amount', ascending=False)  # âœ… Sorted Highest to Lowest

# Initialize Dash app
app = Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Amazon Sales Dashboard", style={'textAlign': 'center'}),
    
    # Summary Cards
    html.Div([
        html.Div([html.H3("Total Orders"), html.P(df.shape[0])], className='card'),
        html.Div([html.H3("Total Revenue"), html.P(f"${df['Amount'].sum():,.2f}")], className='card'),
        html.Div([html.H3("Unique Products"), html.P(df['Category'].nunique())], className='card'),
    ], style={'display': 'flex', 'justify-content': 'space-around'}),
    
    # Charts in a 2-column grid
    html.Div([
        html.Div(dcc.Graph(
            figure=px.histogram(df, x='Size', title='Count of Products By Sizes', color_discrete_sequence=['blue'])
        ), style={'width': '48%', 'display': 'inline-block'}),
        
        html.Div(dcc.Graph(
            figure=px.bar(df_shipping_status, x="Courier Status", y="Amount", 
                          title="Shipping Status Chart",
                          color="Courier Status", 
                          color_discrete_map=color_map)
        ), style={'width': '48%', 'display': 'inline-block'})
    ]),
    
    html.Div([
        html.Div(dcc.Graph(
            figure=px.histogram(df, x='Category', title='Product Sales By Category Histogram Chart', 
                                color_discrete_sequence=['yellow'])
        ), style={'width': '48%', 'display': 'inline-block'}),
        
        html.Div(dcc.Graph(
            figure=px.pie(df, names='B2B', title='Buyer Category Pie Chart', hole=0.3)
        ), style={'width': '48%', 'display': 'inline-block'})
    ]),
    
    html.Div([
        html.Div(dcc.Graph(
            figure=px.pie(df_filtered, names='Fulfilment', 
                          title='Order Fulfilled By Merchant vs Amazon Donut Chart', hole=0.5)
        ), style={'width': '48%', 'display': 'inline-block'}),
        
        html.Div(dcc.Graph(
            figure=px.scatter(df, x='Category', y='Size', title='Available Size Scatter Chart')
        ), style={'width': '48%', 'display': 'inline-block'})
    ]),
    
    html.Div([
        html.Div(dcc.Graph(
            figure=px.line(df.resample('M', on='Date').sum().reset_index(), x='Date', y='Amount', 
                           title='Average Monthly Order Amount Line Chart')
        ), style={'width': '48%', 'display': 'inline-block'}),
        
        html.Div(dcc.Graph(
            figure=px.bar(df_top_states, x='ship-state', y='Amount', title='Top 10 States - Distribution Chart',
                          color='ship-state', color_discrete_sequence=px.colors.qualitative.Pastel)
        ), style={'width': '48%', 'display': 'inline-block'})
    ])
])

# Run app
if __name__ == '__main__':
    app.run(debug=True)
```
